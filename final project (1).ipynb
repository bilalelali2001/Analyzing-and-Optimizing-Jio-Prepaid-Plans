```python
import pandas as pd

# Load the dataset
file_path = 'Jio - Jio Prepaid Planes.csv'  # Update with your actual file path
data = pd.read_csv(file_path)

# Display the first few rows of the dataset
print(data.head())
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import numpy as np

# Step 1: Data Preparation
data_usage = data[['Data_per_day']]

# Step 2: Normalization
scaler = StandardScaler()
data_usage_scaled = scaler.fit_transform(data_usage)

# Step 3: Clustering
# Determine the optimal number of clusters using the elbow method
sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(data_usage_scaled)
    sse.append(kmeans.inertia_)
```

```python
# Plot SSE to find the elbow point
plt.figure(figsize=(10, 5))
plt.plot(range(1, 11, 1), sse)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('SSE')
plt.show()

# From the elbow plot, choose the number of clusters (let's say 3 for this example)
kmeans = KMeans(n_clusters=3, random_state=0)
data['Cluster'] = kmeans.fit_predict(data_usage_scaled)

# Step 4: Visualization
plt.figure(figsize=(10, 5))
sns.scatterplot(x='Data_per_day', y='Profit per customer', hue='Cluster', data=data, palette='viridis')
plt.title('Customer Segmentation Based on Data Usage')
plt.xlabel('Data per Day')
plt.ylabel('Profit per Customer')
plt.show()
```

```python
# Analyze the profitability of different plans
plt.figure(figsize=(10, 5))
sns.barplot(x='Days', y='Profit per customer', data=data)
plt.title('Profitability of Different Plans')
plt.xlabel('Days')
plt.ylabel('Profit per Customer')
plt.xticks(rotation=45)
plt.show()
```

```python
# Calculate the ARPU for each segment
segment_arpu = data.groupby('Cluster')['Profit per customer'].mean().reset_index()
segment_arpu.columns = ['Cluster', 'ARPU']
print(segment_arpu)

# Visualize the ARPU for each segment
plt.figure(figsize=(10, 5))
sns.barplot(x='Cluster', y='ARPU', data=segment_arpu)
plt.title('Average Revenue Per User (ARPU) for Each Segment')
plt.xlabel('Customer Segment')
plt.ylabel('ARPU')
plt.show()
```

```python
import pandas as pd
import numpy as np

# Assuming churn data is not available, let's create a hypothetical Churn column
# For simplicity, let's assume customers with profit per customer < 100 are more likely to churn
data['Churn'] = np.where(data['Profit per customer'] < 100, 1, 0)

# Display the first few rows to verify the new column
print(data.head())
```

```python
# Select relevant features
features = ['Days', 'Price', 'Price/Day', 'For 365 Days', 'Data_per_day']
X = data[features]
y = data['Churn']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, roc_auc_score, roc_curve
import matplotlib.pyplot as plt

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Logistic Regression
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
y_pred_logreg = logreg.predict(X_test)

# Decision Tree
dtree = DecisionTreeClassifier()
dtree.fit(X_train, y_train)
y_pred_dtree = dtree.predict(X_test)

# Random Forest
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
```

```python
# Function to evaluate model performance
def evaluate_model(y_test, y_pred):
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    return accuracy, precision, recall

# Evaluate Logistic Regression
acc_logreg, prec_logreg, recall_logreg = evaluate_model(y_test, y_pred_logreg)
print(f"Logistic Regression - Accuracy: {acc_logreg}, Precision: {prec_logreg}, Recall: {recall_logreg}")

# Evaluate Decision Tree
acc_dtree, prec_dtree, recall_dtree = evaluate_model(y_test, y_pred_dtree)
print(f"Decision Tree - Accuracy: {acc_dtree}, Precision: {prec_dtree}, Recall: {recall_dtree}")

# Evaluate Random Forest
acc_rf, prec_rf, recall_rf = evaluate_model(y_test, y_pred_rf)
print(f"Random Forest - Accuracy: {acc_rf}, Precision: {prec_rf}, Recall: {recall_rf}")

# ROC-AUC Scores
roc_auc_logreg = roc_auc_score(y_test, logreg.predict_proba(X_test)[:, 1])
roc_auc_dtree = roc_auc_score(y_test, dtree.predict_proba(X_test)[:, 1])
roc_auc_rf = roc_auc_score(y_test, rf.predict_proba(X_test)[:, 1])
print(f"Logistic Regression - ROC-AUC: {roc_auc_logreg}")
print(f"Decision Tree - ROC-AUC: {roc_auc_dtree}")
print(f"Random Forest - ROC-AUC: {roc_auc_rf}")

# Plot ROC Curve
plt.figure(figsize=(10, 6))
fpr_logreg, tpr_logreg, _ = roc_curve(y_test, logreg.predict_proba(X_test)[:, 1])
fpr_dtree, tpr_dtree, _ = roc_curve(y_test, dtree.predict_proba(X_test)[:, 1])
fpr_rf, tpr_rf, _ = roc_curve(y_test, rf.predict_proba(X_test)[:, 1])

plt.plot(fpr_logreg, tpr_logreg, label='Logistic Regression (area = %0.2f)' % roc_auc_logreg)
plt.plot(fpr_dtree, tpr_dtree, label='Decision Tree (area = %0.2f)' % roc_auc_dtree)
plt.plot(fpr_rf, tpr_rf, label='Random Forest (area = %0.2f)' % roc_auc_rf)

plt.plot([0, 1], [0, 1], color='navy', linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import minimize

# Define a simple profit function for optimization
def profit_function(price, price_per_day, fixed_costs):
    return price - price_per_day * fixed_costs

# Define a function to be minimized (negative profit for maximization)
def objective(price, price_per_day, fixed_costs):
    return -profit_function(price, price_per_day, fixed_costs)

# Assume fixed costs and initial price per day
fixed_costs = 100  # example fixed costs
initial_price_per_day = 10  # example price per day

# Set  bounds based on the dataset
min_price_per_day = data['Price/Day'].min()
max_price_per_day = data['Price/Day'].max()
bounds = [(min_price_per_day, max_price_per_day)]

# Optimize the price per day to maximize profit
result = minimize(objective, x0=[initial_price_per_day], args=(initial_price_per_day, fixed_costs), bounds=bounds, method='L-BFGS-B')
optimal_price_per_day = result.x[0]
print(f"Optimal Price per Day: {optimal_price_per_day}")

# Conduct Sensitivity Analysis
price_range = np.linspace(min_price_per_day, max_price_per_day, 100)
profits = [profit_function(price, initial_price_per_day, fixed_costs) for price in price_range]

plt.figure(figsize=(10, 5))
plt.plot(price_range, profits)
plt.title('Sensitivity Analysis of Price on Profit')
plt.xlabel('Price per Day')
plt.ylabel('Profit')
plt.axvline(optimal_price_per_day, color='red', linestyle='--', label=f'Optimal Price: {optimal_price_per_day}')
plt.legend()
plt.show()

# Recommendations
print("Recommendations:")
print(f"1. Adjust the price per day to {optimal_price_per_day:.2f} to maximize profit.")
print("2. Monitor customer retention to ensure that the price change does not negatively impact customer satisfaction.")
print("3. Implement targeted marketing strategies for different customer segments based on their price sensitivity and usage patterns.")
```
